### Question 1
Définir une variable schema_movies

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
schema_movies = {'ID':'i','TITLE':'U','YEAR':'i','RUNTIME':'i','RANK':'i'}
```

### Question 2
1. importer module pandas
2. charger le fichier CSV movies et afficher le dataFrame obtenu

```python
movies = pd.read_csv("/home/myriem/Documents/M1-ISD/programmationPython/tp_note/movies.csv", delimiter=";", dtype=schema_movies)
movies
```

### Question 3  
[Statistiques]
* Nombre de lignes
* Année la plus grande
* Moyenne des durées

```python
nb_lignes = movies.shape[0]
annee_max = movies['YEAR'].max()
moy_duree = movies['RUNTIME'].mean()

print("Le nombre de lignes = ",nb_lignes)
print("L'annee la plus grande : ",annee_max)
print("Moyenne des durees = ",moy_duree)
```

### Question 4 
[Filtrage]
* Les films dont le rang est inférieur à 20
* Les films dont la durée est supérieure à la moyenne des durées
* Les films dont le titre contient ’Wars’

```python
rang_inf_20 = movies[movies['RANK'] < 20]

print("Les films dont le rang est inferieur à 20 = ",rang_inf_20)
```

```python
rt_sup_moy = movies[movies['RUNTIME'] > movies['RUNTIME'].mean()]

print("Les films dont la duree est superieure a la moyenne des durees : ",rt_sup_moy)
```

```python
title_wars = movies[movies['TITLE'].str.contains("Wars")]

print("Les films dont le titre contient 'Wars' : ",title_wars)
```

# Partie 6 : Matplotlib

### Question 7
1. Nombre de films par année et afficher le résultat
2. Copie du dataFrame movies
3. sur cette copie mettre à jour la colonne 'YEAR' en arrondissant chaque année à la décennie
4. groupby par année et compter le nombre de films
5. Tracage d'histogramme

```python
nb_film_annee = movies.groupby('YEAR', as_index=False).size()
nb_film_annee
```

```python
data_frame_movies = movies.copy()
data_frame_movies['YEAR'] = (data_frame_movies['YEAR']//10)*10
data_frame_movies
```

```python
data_grouped = data_frame_movies.groupby('YEAR', as_index=False).size()
data_grouped
```

```python
data_grouped.plot.bar(x='YEAR',y='size')
```

### Question 8
affichage moyenne des rangs des films par decennie

```python
data_mean_rank = data_frame_movies.groupby('YEAR', as_index=False)['RANK'].mean()
data_mean_rank
```

```python
data_mean_rank.plot.bar(x='YEAR',y='RANK')
```
